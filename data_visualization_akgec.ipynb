```python
#1. pyplot  --> Python --
#2. pylab  --> MATLAB

import matplotlib.pyplot as plt #from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Line plot (plt.plot(x, y ,formatting options = [color][marker][line]))
# 'r0'(red markers) 'ro-'(red line markers) 'ro--'(red dashed line markers) 'r--'(red dashed line)
plt.plot([1,2,3,4],[2,3,4,5],'ro--')
plt.show()
```

```python
plt.plot([10,20,30,40],'ro')
plt.show()
```

```python
#bar chart
x = np.array([1,2,3,4,5])
plt.bar(x,[10,20,33,46,78], width = 0.2, color = 'red')
plt.bar(x + 0.2,[25,35,56,58,76], width = 0.2)
plt.xlabel('Rank')
plt.ylabel('Percentage')
plt.title('Rank vs Percentage')
plt.legend(['Female','Male'])
plt.show()
```

```python
#histograms
x = np.random.randint(1,4,1000)
plt.hist(x, bins = [-0.5,0.5,1.5])
```

```python
#boxplot

#first quartile
x = np.random.randint(1,100,1000)
x = np.append(x, -242)
plt.boxplot(x, notch = 1)
plt.show()
```

```python
#pie chart

x = [23, 45, 56, 65, 43, 121]
label = ['google','facebook','hyundai','bmw','apple','benz']
plt.pie(x, labels = label, autopct = '%.4f')
plt.show()
```

## Seaborn

```python
import numpy as np    #for scientific computation
import pandas as pd   #for data - preprocessing
import matplotlib.pyplot as plt  #for basic plotting
import seaborn as sns            #for advanced plotting
```

```python
x = [1,2,3,4]
y = [2,3,4,5]
#lineplot
sns.lineplot(x, y)
plt.show()
```

```python
#bar plot
sns.barplot(x, y)
```

```python
#kdeplot (kernel density estimation)
x=np.random.random(100)*100
sns.kdeplot(x)
plt.show()
```

```python
#boxplot
sns.boxplot(x)
```

```python
#distplot (kde or histogram)
sns.distplot(x, kde = False, bins = [0,25,35,45])
```

```python
x=np.random.randint(10,size=10)
y=np.random.randint(10,size=10)
hue=['Male','Female','Male','Male','Female','Female','Male','Female','Male','Male']
ax=sns.scatterplot(x, y, hue=hue)
ax.set(xlabel='X-Axis', ylabel='Y-Axis')
ax.set(title='Scatter Plot')
```
